```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV, train_test_split

import warnings
warnings.filterwarnings("ignore")

np.random.seed(0)
%matplotlib inline
```

# 1. Read and Preprocess data

```python
df = pd.read_csv('./Hitters.csv')
# drop rows with NA
df = df.dropna()

# drop non-numeric columns and 'Salary'
x = df.drop(df.columns[[0]], axis=1)
x = x.drop(columns=['League', 'Division', 'Salary', 'NewLeague']) 

# normalization
# x = (x-x.mean())/x.std()
# x = (x-x.min())/(x.max()-x.min())

y = df['Salary']

# test_size == the one in video 06
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
```

# 2. Visualize the coefficient trajectories(as a function of regularization)

```python
alphas = np.logspace(-5, 3, 100)

coefs = []
for a in alphas:
    ridge = Ridge(alpha=a, normalize=True)
    ridge.fit(x_train, y_train)
    coefs.append(ridge.coef_)

plt.rcParams["figure.figsize"] = (20,10)
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefs')
plt.show()
```

# 3. Use CV to find the optimal value of the regularization penality

```python
r = Ridge(normalize=True)
param_grid = {'alpha': [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]}
cv = GridSearchCV(r, param_grid, cv=20)
cv.fit(x_train, y_train)
print(cv.best_params_)
print('train accuracy:\t', cv.score(x_train, y_train))
print('test accuracy:\t', cv.score(x_test, y_test))
```
