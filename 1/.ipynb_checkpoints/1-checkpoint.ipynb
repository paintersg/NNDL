```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV, train_test_split

import warnings
warnings.filterwarnings("ignore")

np.random.seed(0)
%matplotlib inline
```

# 1. LASSO

## 1.1 Read and preprocess data

```python
df = pd.read_csv('./Hitters.csv')
# drop rows with NA
df = df.dropna()

# drop non-numeric columns and 'Salary'
x = df.drop(df.columns[[0]], axis=1)
x = x.drop(columns=['League', 'Division', 'Salary', 'NewLeague']) 

# normalization
# x = (x-x.mean())/x.std()
# x = (x-x.min())/(x.max()-x.min())

y = df['Salary']

# test_size == the one in video 06
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
```

## 1.2 Visualize the coefficient trajectories(as a function of regularization)

```python
alphas = np.logspace(-3, 3, 100)

coefs = []
for a in alphas:
    lasso = Lasso(alpha=a, normalize=True)
    lasso.fit(x_train, y_train)
    coefs.append(lasso.coef_)

plt.rcParams["figure.figsize"] = (20,10)
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefs')
plt.show()
```

## 1.3 Comment on which are the final three predictors that remain in the model

It's shown in the plot that there are only three non-zero coefs left(blue, red, purple) when alpha equals some value between 9 and 20.
The predictors corresponding to those three non-zero coefs are the final three predictors.

```python
lasso = Lasso(alpha=10, normalize=True)
lasso.fit(x_train, y_train)
lasso.coef_
```

So the final three predictors are column 3,4,10(starting from 0). The name of these columns is shown below.

```python
print(x.columns[3], x.columns[4], x.columns[10])
```

## 1.4 Use cross-validation to find the optimal value of the regularization penality

```python
l = Lasso(normalize=True)
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
cv = GridSearchCV(l, param_grid, cv=20)
cv.fit(x_train, y_train)
print(cv.best_params_)
print('train accuracy:\t', cv.score(x_train, y_train))
print('test accuracy:\t', cv.score(x_test, y_test))
```

**optimal alpha is 1**

## 1.5 How many predictors are left in that model?

```python
lasso2 = Lasso(alpha=1, normalize=True)
lasso2.fit(x_train, y_train)
lasso2.coef_
```

There are 8 non-zero coefs left, but 0.0077 and 0.0621 are quite close to 0. So I think there are 6 predictors left in that model.

## 1.6 Predict salary

```python
prediction = lasso2.predict(x_test)
print(prediction)
```

# 2. Ridge

## 2.1 Visualize the coefficient trajectories(as a function of regularization)

```python
alphas = np.logspace(-5, 3, 100)

coefs = []
for a in alphas:
    ridge = Ridge(alpha=a, normalize=True)
    ridge.fit(x_train, y_train)
    coefs.append(ridge.coef_)

plt.rcParams["figure.figsize"] = (20,10)
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefs')
plt.show()
```

## 2.2 Use cross-validation to find the optimal value of the regularization penality

```python
r = Ridge(normalize=True)
param_grid = {'alpha': [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]}
cv = GridSearchCV(r, param_grid, cv=20)
cv.fit(x_train, y_train)
print(cv.best_params_)
print('train accuracy:\t', cv.score(x_train, y_train))
print('test accuracy:\t', cv.score(x_test, y_test))
```

# 3. Short answer

## 3.1 Bias-variance tradeoff
**Bias:**
Bias of an instance is the difference between predicted values and true values.
Bias of a model is the average bias of several instances generated by the model.

But for models, low bias doesn’t mean high accuracy. If a model with low bias has high variance, it could generate instances with high bias, while the average value of these bias is still low.

**Variance:**
Variance describes the stability of a model. The instances generated by a model with low variance are similar to each other, while instances generated by a model with high variance varies largely.

**Bias-variance tradeoff:**
To lower bias, we need the predictions of each instance to be more close to true values. To make this possible, the model should be more complicated so that it will have a more powerful ability to approximate true values. But as a result, the instances generated by this model can be very twisted and sensitive to small fluctuations in training set. So for a given point, the predicted values of different instances can be quite different, that is to say these instances can be quite different. So the model will have a high variance.

To lower variance, we need to make instances less sensitive to small fluctuations. So they can’t be too twisted. So the model should be more simple, which will weaken its ability to approximate true values. As a result, the bias will be higher.

## 3.2 What role does regularization play in this tradeoff
Regularization prevents instances from becoming too complicated and twisted by keeping coefficients small. If some coefficient is equal to zero, then the corresponding term can be ignored, which will simplify the instance. As a result, there will be less difference between the instances generated at different times and the variance of the model will be reduced.

## 3.3 Describe models of high/low bias and variance
In both LASSO and Ridge, the best alpha is 1. When alpha is small, the absolute values of coefficients are large, which makes the instances complicated. In this case, the bias of the model can be low and the variance can be high. When alpher is large, the absolute values of coefficients are small, which makes the instances simple. In this case, the bias of the model can be higher and the variance can be lower.
Regularization help the tradeoff to achieve a balance.
